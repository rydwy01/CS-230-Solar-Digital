```python
#my attempt at following instructions at https://albumentations.ai/docs/getting_started/image_augmentation/ is in the code below
import albumentations as A
import cv2
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
#from IPython.display import Image
```

```python
#some examples of augmentations that could be relevant for our project.  A LOT more on the documentation
transform = A.Compose([
    A.RandomCrop(width=2048, height=2048),
    A.HorizontalFlip(p=0.99),
    A.VerticalFlip(p=1),
    A.RandomBrightnessContrast(p=0.99),
    A.RGBShift(r_shift_limit=100, g_shift_limit=100, b_shift_limit=100, always_apply=False, p=0.99)
])
```
